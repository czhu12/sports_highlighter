```python
from collections import OrderedDict

#goal is to organize first by date, then by number of people who visit each city on a particular day
#key must be date, but key must also be city within that date. store users as a set to prevent duplicates and O(1) time access


def pop(N, arr):
    obj = dict()
    
    for item in arr:
        user_id, listing_id, city, start_date, end_date = item.split(',')
        curr_date = s_to_date(start_date)
        
        #until we reach the end of the date
        while (curr_date <= s_to_date(end_date)):
            curr_date_s = date_to_s(curr_date)
            #add user and city in obj
            if curr_date_s in obj:
                if city in obj[curr_date_s]:
                    obj[curr_date_s][city].add(user_id)
                else:
                    obj[curr_date_s][city] = set()
                    obj[curr_date_s][city].add(user_id)
            else:
                obj[curr_date_s] = dict()
                new_s = set()
                new_s.add(user_id)
                obj[curr_date_s][city] = new_s
            
            curr_date = add_one_day(curr_date)
    
    
    obj = OrderedDict(sorted(obj.items()))
    
    
    res = []

    for date in obj:
        sorted_set = sorted(sorted(obj[date]), key=lambda city: len(obj[date][city]), reverse=True)
        top_n = sorted_set[0:N]
        s = build_string(obj, date, top_n)
        res.append(s)
        
    return res


def add_one_day(date):
    return date + timedelta(days=1)

def date_to_s(parsed_date):
    return datetime.strftime(parsed_date, "%Y-%m-%d")

def s_to_date(date):
    return datetime.strptime(date, "%Y-%m-%d")

def build_string(obj, date, arr):
    s = ""
    s += date
    s += ","
    for i in range(len(arr)):
        city = arr[i]

        s += city
        s += ":"
        s += str(len(obj[date][city]))
        if (i < len(arr) - 1):
            s += ","
    return s


```

```python
#their sol
from datetime import datetime
from datetime import timedelta

def parseDateString(dateStr):
    return datetime.strptime(dateStr, '%Y-%m-%d')

def dateToString(date):
    return date.strftime('%Y-%m-%d')

def popular(num, requests):
    dateStrToCity = dict()
    
    for request in requests:
        user_id, listing_id, city, start_date, end_date = request.split(',')
        date = parseDateString(start_date)
        
        while date <= parseDateString(end_date):
            dateStr = dateToString(date)
            if dateStr not in dateStrToCity:
                dateStrToCity[dateStr] = dict()
            if city not in dateStrToCity[dateStr]:
                dateStrToCity[dateStr][city] = set()
                
            dateStrToCity[dateStr][city].add(user_id)
            date += timedelta(days=1)
            
    ret = []
    
    
    for date in dateStrToCity:
        def scoreCity(city):
            return len(dateStrToCity[date][city])

        def compareCities(city1, city2):
            s1, s2 = scoreCity(city1), scoreCity(city2)

            if s1 != s2:
                return s2-s1
            if city1 < city2:
                return -1
            return 1
        
        s = date + ','
        
        cities=sorted([city for city in dateStrToCity[date]], cmp=compareCities)
        for i, city in enumerate(cities):
            s += city
            s += ':'
            s += str(scoreCity(city))
            s += ','
            if i + 1 == num:
                break
        s = s[:-1]
        
        ret.append(s)
        
    ret = sorted(ret)
    return ret
```
